# Plot a bar graph of male vs female buyers participated in the sales.

```python
df=pd.DataFrame(data)
gender=df['Buyer Gender'].value_counts()

gender.plot(kind='bar',figsize=(10,8))
plt.title('male vs female')
plt.show()
```

# Find the top 5 cars based on their sales price.

```python
df=pd.DataFrame(data)
car=df['Model'].value_counts()[:5]
sp=df.sort_values(by='Sale Price').value_counts()[:5]
sp.plot(kind='bar',figsize=(10,8))

plt.xticks()
plt.yticks()
plt.show()
print(sp)
```
